# Enunciado

Escribe un programa que evalúe las calificaciones de un grupo de estudiantes para determinar su promedio final. El programa debe filtrar las notas no válidas, que incluyen números fuera del rango de 0 a 100 y valores que no son numéricos. Las notas válidas deben ser utilizadas para calcular el promedio de cada estudiante. Si una nota es inválida, el programa deberá notificar el error y mostrar el nombre del estudiante junto con la descripción del error. Finalmente, el programa debe imprimir el nombre del estudiante y su calificación promedio basada solo en las notas válidas.


Ejemplo Entrada:

``` python
students = [
    {"nombre": "José", "notas": [85, 90, 78, -10]},
    {"nombre": "Carlos", "notas": [88, 72, 94, 102]},
    {"nombre": "María", "notas": [95, 85, 87, "A++"]}
]

```

Salida esperada:


``` python

Error en la nota de José: Nota inválida -10
José: Calificación final = 84.33
Error en la nota de Carlos: Nota inválida 102
Carlos: Calificación final = 84.67
Error en la nota de María: Nota inválida A++
María: Calificación final = 89.00

```

## Ayudita :)


*Entrada*: Una lista de diccionarios, donde cada diccionario representa a un estudiante con su nombre y una lista de notas.

*Proceso*:
El programa itera sobre cada estudiante.
Extrae el nombre y las notas del estudiante.
Filtra y valida cada nota para asegurarse de que es un número dentro del rango de 0 a 100.
Si encuentra una nota inválida, muestra un mensaje de error especificando la nota y el estudiante asociado.
Calcula el promedio de las notas válidas y muestra el nombre del estudiante junto con su calificación final calculada.

*Salida*: Mensajes de error para notas inválidas y el promedio de calificaciones válidas para cada estudiante.


````python
students = [
    {"nombre": "José", "notas": [85, 90, 78, -10]},
    {"nombre": "Carlos", "notas": [88, 72, 94, 102]},
    {"nombre": "María", "notas": [95, 85, 87, "A++"]}
]

for student in students:
    nombre = student["nombre"]
    notas = student["notas"]

    print("nombre: ",nombre, " su notas son: ",notas )
````

Podría obtenerse una salida más limpia eliminando el mensaje que especifica el tipo de error que hubo con la nota, pero quise aprovechar para usar Try-Except y que quede evidenciado el error.

````python
students = [
    {"nombre": "Pedro", "notas": [20, 100, 89, 87]},
    {"nombre": "José", "notas": [85, 90, 78, -10]},
    {"nombre": "Carlos", "notas": [88, 72, 94, 102]},
    {"nombre": "María", "notas": [95, 85, 87, "A++"]},
    {"nombre": "Carla", "notas": [100, 45, 65, "91"]},
    {"nombre": "Guillermina", "notas": []},
    {"nombre": "Ana", "notas": ["A", "B", "C"]},
    {"nombre": "Juana", "notas": ["89", 74, "B", "107", 109, 95]}
]

for student in students:
    nombre = student["nombre"]
    notas = student["notas"]
    suma = 0
    promedio = 0
    notas_validas = []

    for i in notas:
      tipoDeDato = type(i)


      if (tipoDeDato == int or tipoDeDato == float) and (0 <= i <= 100):
            suma += i
            notas_validas.append(i)

      elif (tipoDeDato == int or tipoDeDato == float) and (i < 0 or i > 100):
        print("Error en la nota de ", nombre, ". Nota inválida: ", i)

      elif (tipoDeDato == str):

        try:
          nota= int(i)
          if (0 <= nota <= 100):
            suma = suma + nota
            notas_validas.append(nota)
          else:
            print("Error en la nota de ", nombre, ". Nota inválida. ", i)
        except Exception as e:
          print("Error en la nota de ", nombre, ". Nota inválida: ", i)
          print("El error fue {}".format(e))

    if len(notas_validas) > 0:
      promedio = suma/len(notas_validas)

    else:
      print("No hay notas válidas para", nombre)

    print(nombre, ": ", "Calificación final = ", promedio)
````
